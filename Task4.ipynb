```python
# Problem 1 :
valid_plans = [199, 299, 399, 599]

while True:
    # Get input from the user
    amount = int(input("Enter recharge amount (₹): "))
    
    # Check the rules
    if amount < 50:
        print("Invalid: Recharge amount must be ₹50 or more. Try again.\n")
    elif amount in valid_plans:
        print(f"Success! Your mobile has been recharged with ₹{amount}.\n")
        break  # Exit the loop to move to the next problem
    else:
        print("Invalid plan. Available plans are ₹199, ₹299, ₹399, ₹599. Try again.\n")

# =========================================================

# Problem 2 :
inventory = {
    "Laptops": 10,
    "Mice": 45,
    "Keyboards": 12,
    "Monitors": 20
}

# Loop through the dictionary items
for product, stock in inventory.items():
    if stock < 15:
        print(f"Reorder Alert: {product} (Stock: {stock})")
    else:
        print(f"Stock OK: {product} (Stock: {stock})")
print() # Blank line for spacing

# =========================================================

# Problem 3 :
student_marks = [45, 60, 55, 70, 40]
total_marks = 0

# Use a loop to calculate the total sum of the marks
for mark in student_marks:
    total_marks += mark

# Calculate average and display result
average = total_marks / len(student_marks)
print(f"Average Marks: {average}")

if average >= 50:
    print("Result: Pass\n")
else:
    print("Result: Fail\n")

# =========================================================

# Problem 4 :
while True:
    # Get trip details
    km = float(input("Enter distance in kilometers: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    
    # Calculate base fare (₹50 base + ₹12 per km)
    fare = 50 + (km * 12)
    
    # Add 25% if it is a peak hour
    if peak == "yes":
        fare = fare + (fare * 0.25)
        
    print(f"Estimated Cab Fare: ₹{fare}")
    
    # Check if the user wants to calculate another fare
    retry = input("\nCalculate another fare? (yes/no): ").lower()
    if retry != "yes":
        print("Exiting Cab Estimator...\n")
        break  # Stop the loop to move to the next problem

# =========================================================

# Problem 5 :
# "P" = Present, "A" = Absent
attendance_record = ["P", "P", "A", "P", "P", "A", "P", "P", "P", "P"]
present_days = 0

# Loop to count present days
for status in attendance_record:
    if status == "P":
        present_days += 1

# Calculate percentage
percentage = (present_days / len(attendance_record)) * 100
print(f"Attendance Percentage: {percentage}%")

# Check eligibility
if percentage >= 75:
    print("Status: Eligible\n")
else:
    print("Status: Not Eligible\n")


# =========================================================

# Problem 6 :
password = "User@1234"
print(f"Checking password: {password}")

# Rule 1: Check minimum length
if len(password) < 8:
    print("Weak: Password must be at least 8 characters long.")
else:
    has_digit = False
    has_special = False
    valid_specials = "@#$"
    
    # Loop over the string to check for digits and special characters
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in valid_specials:
            has_special = True
            
    # Evaluate the results
    if has_digit and has_special:
        print("Strong: Password is valid.")
    elif not has_digit:
        print("Weak: Password must contain at least one digit.")
    else:
        print("Weak: Password must contain at least one special character (@, #, $).")
```
