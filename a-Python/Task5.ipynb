```python
# --------------------------------------------------
# Problem : 1

def analyze_engagement(likes_data):
    total_likes = 0
    for likes in likes_data:
        total_likes += likes
        
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
        
    print("--- Problem 1 Output ---")
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}\n")

analyze_engagement([250, 300, 150, 350])


# --------------------------------------------------
# Problem : 2

def monitor_stock(stock_level):
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
        
    print("--- Problem 2 Output ---")
    print(f"Medicine Stock: {stock_level}")
    print(f"Status: {status}\n")

monitor_stock(6)


# --------------------------------------------------
# Problem : 3

def check_rainfall_adequacy(rainfall_data, required_level):
    total_rainfall = 0
    for rain in rainfall_data:
        total_rainfall += rain
        
    average_rainfall = total_rainfall / len(rainfall_data)
    
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
        
    print("--- Problem 3 Output ---")
    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}\n")

check_rainfall_adequacy([60, 80, 70, 78], 70)


# --------------------------------------------------
# Problem : 4

def detect_duplicate_accounts(account_list):
    unique_accounts = set(account_list)
    
    if len(unique_accounts) < len(account_list):
        duplicates_found = "Yes"
    else:
        duplicates_found = "No"
        
    print("--- Problem 4 Output ---")
    print(f"Duplicate Accounts Found: {duplicates_found}\n")

detect_duplicate_accounts(["user123", "cool_guy", "user123", "admin"])


# --------------------------------------------------
# Problem : 5

def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
        
    print("--- Problem 5 Output ---")
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}\n")

check_appointment_eligibility(21)


# --------------------------------------------------
# Problem : 6

def filter_premium_crops(prices):
    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
            
    print("--- Problem 6 Output ---")
    print(f"Premium Crops: {premium_crops}\n")

filter_premium_crops([1500, 2500, 1800, 3200])


# --------------------------------------------------
# Problem : 7

def check_app_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
        
    print("--- Problem 7 Output ---")
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}\n")

check_app_health(7)


# --------------------------------------------------
# Problem : 8

def validate_transaction(amount):
    daily_limit = 50000
    
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
        
    print("--- Problem 8 Output ---")
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}\n")

validate_transaction(60000)


# --------------------------------------------------
# Problem : 9

def calculate_attendance(attendance_record):
    days_present = 0
    total_days = len(attendance_record)
    
    for day in attendance_record:
        if day == 1:
            days_present += 1
            
    attendance_percentage = (days_present / total_days) * 100
    
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
        
    print("--- Problem 9 Output ---")
    print(f"Attendance Percentage: {attendance_percentage}")
    print(f"Exam Eligibility: {status}\n")

calculate_attendance([1, 1, 1, 1, 0, 1, 1, 1, 0, 1])


# --------------------------------------------------
# Problem : 10

def analyze_electricity_bill(units):
    bill = 0
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
        
    if bill < 500:
        usage = "Low Usage"
    elif 500 <= bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
        
    return bill, usage

print("--- Problem 10 Output ---")
total_bill, usage_status = analyze_electricity_bill(250)
print(f"Total Bill: â‚¹{total_bill}")
print(f"Usage Status: {usage_status}\n")
```
