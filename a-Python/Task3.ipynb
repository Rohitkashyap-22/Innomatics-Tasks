```python
# Problem 1: Employee Performance Bonus Eligibility
# Input data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest score
max_score = max(employees.values())

# Create an empty list and add names that match the max score
top_performers = []
for name, score in employees.items():
    if score == max_score:
        top_performers.append(name)

# Format and display the result
names_str = ", ".join(top_performers)
print(f"Top Performers Eligible for Bonus: {names_str} (Score: {max_score})")

#-----------------------------------------------------------------------------------------------------

#Problem 2: Search Query Keyword Analysis
# Input data
query = "Buy mobile phone buy phone online"

# Convert to lowercase and split into words
query = query.lower()
words = query.split()

# Count the frequency of each word 
word_counts = {}
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Filter for keywords searched more than once
duplicate_keywords = {}
for word, count in word_counts.items():
    if count > 1:
        duplicate_keywords[word] = count

print(duplicate_keywords)

#-----------------------------------------------------------------------------------------------------

#Problem 3: Sensor Data Validation
# Input data
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Create an empty list for valid readings
valid_readings = []

# Loop through the list, checking the index (hour) and the value
for hour, value in enumerate(sensor_readings):
    # Check if the reading is an even number
    if value % 2 == 0:
        valid_readings.append((hour, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)

#-----------------------------------------------------------------------------------------------------

#Problem 4: Email Domain Usage Analysis
# Input data
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

total_emails = len(emails)
domain_counts = {}

# Extract domain and count occurrences
for email in emails:
    domain = email.split('@')[1]
    if domain in domain_counts:
        domain_counts[domain] += 1
    else:
        domain_counts[domain] = 1

# Calculate and display percentages
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    # Convert to integer
    print(f"{domain}: {int(percentage)}%")

#-----------------------------------------------------------------------------------------------------

#Problem 5: Sales Spike Detection
# Input data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average daily sales
total_sales = sum(sales)
average_sales = total_sales / len(sales)

# Calculate the threshold (30% above average)
threshold = average_sales + (average_sales * 0.30)

# Detect and display spikes
for i in range(len(sales)):
    if sales[i] > threshold:
        # i + 1 is used because days start at Day 1, not Day 0
        print(f"Day {i + 1}: {sales[i]}")

#-----------------------------------------------------------------------------------------------------

#Problem 6: Duplicate User ID Detection
# Input data
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_counts = {}

# Count occurrences of each user ID
for uid in user_ids:
    if uid in id_counts:
        id_counts[uid] += 1
    else:
        id_counts[uid] = 1

# Display duplicates
for uid, count in id_counts.items():
    if count > 1:
        print(f"{uid} â†’ {count} times")
```
