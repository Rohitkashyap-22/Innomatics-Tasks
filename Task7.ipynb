Problem 1: Smart Light Controller

```python
class LightController:
    def __init__(self, light_name, status="OFF"):
        self.light_name = light_name
        self.status = status
        
    # methods:-
    def turn_on(self):
        self.status = "ON"
        print(f"{self.light_name} is ON")
    def turn_off(self):
        self.status = "OFF"
        print(f"{self.light_name} is OFF")
    
# Taking Input
light_name = input("Enter Light Name: ")

# Creating Object
light = LightController(light_name)

# Display Output
light.turn_on()
light.turn_off()
```

Problem 2: Employee ID Card System

```python
class Employee:
    # constructor
    def __init__(self, employee_name, employee_id, department):
        self.employee_name = employee_name 
        self.employee_id = employee_id
        self.department = department
    
    # methods:-
    def display_id(self):
        print("Employee ID Card")
        print(f"Name: {self.employee_name}")    
        print(f"ID: {self.employee_id}")    
        print(f"Department: {self.department}")    

# taking input
name = input("Enter the employee name: ")
ID = input("Enter the employee id: ")
Department = input("Enter the employee department: ")

# creating object
employee = Employee(name, ID, Department)

# display output
employee.display_id()
```

Problem 3: Mobile Contact Record

```python
class Contact:
    # constructor
    def __init__(self, contact_name, contact_number):
        self.contact_name = contact_name
        self.contact_number = contact_number
    
    # methods
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.contact_name}")
        print(f"Phone: {self.contact_number}")

# input taking
name = input("Enter the name: ")
contact = input("Enter the contact number: ")

# function call and object creation
Person_contact = Contact(name, contact)
Person_contact.display_contact()
```

Problem 4: Product Price Tag Generator

```python
class ProductTag:
    # constructor
    def __init__(self, product_name, product_price):
        self.product_name = product_name
        self.product_price = product_price
    
    # methods
    def tag_generator(self):
        print(f"Product: {self.product_name}")
        print(f"Price: Rs.{self.product_price}")                               
    
#take input
name = input("Enter the product name: ")
price = int(input("Enter the price of the product: "))

#fxn call and obj creation
product = ProductTag(name, price)
product.tag_generator() 
```

Problem 5: Movie Rating Display System

```python
class MovieSystem:
    def __init__(self, movie_name, movie_rating):
        self.movie_name = movie_name
        self.movie_rating = movie_rating
    
    def display(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.movie_rating} / 5")

#take inputs
name = input("Enter the name of the movie: ")
rating = float(input("Eneter the rating of the movie: "))

# create object
movie = MovieSystem(name, rating)
movie.display()
```
