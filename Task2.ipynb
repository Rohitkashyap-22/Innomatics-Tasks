```python
# Problem 1: Unique Words in a Sentence
# Taking input sentence
sentence = input("Enter sentence: ")

# Split into words
words = sentence.split()

# Convert list to set to remove duplicates
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)

# ------------------------------------------------------------------------------------------

# Problem 2: Highest Salary from Employee Data
# Given dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Finding employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Displaying output result
print("Highest Salary:", highest_employee, "-", highest_salary)

# ------------------------------------------------------------------------------------------

# Problem 3: Find Maximum and Minimum Values
# Taking input from user and convert into list of integers
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

# Assuming first element is both maximum and minimum
maximum = numbers[0]
minimum = numbers[0]

# Loop through each number in the list
for num in numbers:
    # If current number is greater than maximum, update maximum
    if num > maximum:
        maximum = num
    
    # If current number is smaller than minimum, update minimum
    if num < minimum:
        minimum = num

# Display results
print("Max:", maximum)
print("Min:", minimum)

# ------------------------------------------------------------------------------------------

# Problem 4: Count Products Above a Price Threshold
# Take product prices as input
prices = list(map(int, input("Enter prices separated by space: ").split()))

count = 0  # Counter to store how many prices are above 1000

for price in prices:
    if price > 1000:
        count += 1  # Increase counter

# Display result
print("Products above 1000:", count)

# ------------------------------------------------------------------------------------------

# Problem 5: Calculate Attendance Percentage
# Take attendance input (P for Present, A for Absent)
attendance = input("Enter P or A separated by space: ").split()

present = 0  # Counter for present days

for day in attendance:
    # Convert to uppercase to handle small/large letters
    if day.upper() == "P":
        present += 1

# Calculate percentage
percentage = (present / len(attendance)) * 100

# Display result
print("Attendance Percentage:", percentage)


# ------------------------------------------------------------------------------------------

# Problem 6: Remove Duplicate Phone Numbers
# Take phone numbers as input
phone_numbers = list(map(int, input("Enter phone numbers separated by space: ").split()))

# Convert list to set. set will automatically removes duplicates
unique_numbers = set(phone_numbers)

# Display unique numbers
print("Unique phone numbers:", unique_numbers)

# ------------------------------------------------------------------------------------------

# Problem 7: Count Character Frequency
# Take string input
text = input("Enter a string: ")

frequency = {}                      # Create empty dictionary to store frequency

for char in text:                   # Loop through each character in string
    if char in frequency:
        frequency[char] += 1        # If character already exists in dictionary, increase count
    else:
        frequency[char] = 1         # If character not present, add it with count 1

# Display result
print("Character Frequency:", frequency)


# ------------------------------------------------------------------------------------------

# Problem 8: Convert List to Tuple

# Take numbers as input
numbers = list(map(int, input("Enter numbers separated by space: ").split()))

# Convert list into tuple
numbers_tuple = tuple(numbers)

# Display tuple
print("Tuple:", numbers_tuple)

# ------------------------------------------------------------------------------------------

# Problem 9: Check if a Key Exists in a Dictionary
employees = {}  # Empty dictionary

# taking input of no. of employees
n = int(input("How many employees? "))

# Take employee name and salary as input
for i in range(n):
    name = input("Enter employee name: ")
    salary = int(input("Enter salary: "))
    employees[name] = salary  # Store in dictionary

# Ask which employee to check
key = input("Enter name to check: ")

# Check if key exists
if key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")

# ------------------------------------------------------------------------------------------

# Problem 10: Calculate Average Marks
# Take marks input
marks = list(map(int, input("Enter marks separated by space: ").split()))

# Calculate total using sum()
total = sum(marks)

# Calculate average
average = total / len(marks)

# Display result
print("Average Marks:", average)
```
