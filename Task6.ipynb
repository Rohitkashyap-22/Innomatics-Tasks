# Problem 1 : Smart Parking Lot Management System

```python
def parking_system(vehicle_logs, capacity):
    peak_parking = 0
    current_count = 0
    for logs in vehicle_logs:
        clean_log = logs.strip() #remove any extra space
        
        if clean_log == 'IN':
            current_count += 1
        else:
            current_count = max(0, current_count-1)

        # updating peak parking
        if peak_parking < current_count:
            peak_parking = current_count
    
    # checing status of parking
    if current_count < capacity:
        status = "Available"
    else: 
        status = "Parking Full !!!"
    # results printing
    print( f"Currently parked vehicles: {current_count}")
    print( f"Peak parking usage: {peak_parking}")
    print( f"Parking status: {status}")

# taking inputs
capacity = int(input("Enter the maximum capacity of the parking lot: "))
vehicle_logs = input("Enter logs: ").upper().split(",")

parking_system(vehicle_logs, capacity)
```

# Problem 2:Online Food Delivery Time Estimator

```python
def ETA(distance, traffic, weather):

    #Set default delays to 0 (Protect against typos)
    traffic_delay = 0
    weather_delay = 0

    # traffic conditions check
    if traffic == "Low":
        traffic_delay = 0
    elif traffic == "Medium":
        traffic_delay = 5
    elif traffic == "High":
        traffic_delay = 10
    
    # weather conditions check
    if weather == "Clear":
        weather_delay = 0
    if weather == "Rainy":
        weather_delay = 5

    time = distance*5 + traffic_delay + weather_delay
    print(f"Estimated Delivery Time: {time} minutes")

# taking inputs
distance = float(input("Enter the distance of delivery: " ))
traffic = input("Enter the traffic conditions as Low/Medium/High: ").strip().capitalize()
weather = input("Enter the weather conditions as Rainy/Clear: ").strip().capitalize()

ETA(distance, traffic, weather)
```

# Problem 3: Movie Theatre Seat Occupancy Analyzer

```python
def Seat_Occupancy(total_Seats, booked_seats):
    occupied_count = 0                         #occupied seats
    for seat in booked_seats:
        if seat == 1:
            occupied_count += 1
    
    occupancy = (occupied_count / total_Seats) * 100
    if occupancy == 100:
        status = "Housefull" 
    elif occupancy >= 80:
        status = "Almost Full" 
    else:
        status = "Available"

    print("occupancy: ",occupancy)
    print("Show Status: ",status)

    if occupancy >= 80:
        print("Suggestion: High demand! Consider opening additional shows.")
    
# input taking
total_Seats = int(input("enter the number of total seats: "))
seats = input("enter the seats occupancy matrix: ").split(",")
booked_seats = [int(seat.strip()) for seat in seats]

Seat_Occupancy(total_Seats, booked_seats)
```

# Problem 4: Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    # Calculate the average
    total_load = sum(cpu_readings)
    average_load = total_load / len(cpu_readings)
    
    # Classify the load
    if average_load < 50:
        status = "Normal"
    elif average_load <= 80:  # Covers 50% to 80%
        status = "Warning"
    else:                     # Covers anything above 80%
        status = "Critical"
        
    # Output
    print(f"\nAverage CPU Load: {average_load:.0f}%")
    print(f"Server Status: {status}")

# Taking Input
raw_readings = input("Enter CPU readings separated by commas (e.g. 45,60,70): ").split(",")
readings_list = [int(reading.strip()) for reading in raw_readings]

classify_server_load(readings_list)
```

# Problem 5: Smart Classroom Resource Usage Monitor

```python
def monitor_resources(usage_dict):
    overused_resources = []
    threshold = 8  # Anything over 8 hours triggers the alert
    
    # .items() lets us loop through both the Name and the Hours at the same time
    for resource_name, hours_used in usage_dict.items():
        if hours_used > threshold:
            overused_resources.append(resource_name)
            
    # Check if list of overused resources has anything in it
    if len(overused_resources) > 0:
        alert = "Yes"
        # .join() cleanly combines a list into a string
        final_resources = ", ".join(overused_resources)
    else:
        alert = "No"
        final_resources = "None"
        
    print(f"\nOverused Resources: {final_resources}")
    print(f"Energy Alert: {alert}")

# Defining the Dictionary 
# In Python, dictionaries use { } with "Key": Value pairs
classroom_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(classroom_usage)
```

# Problem 6: Online Event Registration Capacity Controller

```python
def manage_registration(capacity, total_incoming_requests):
    confirmed = 0
    waitlisted = 0
    
    # We loop exactly 'total_incoming_requests' times
    for _ in range(total_incoming_requests):
        if confirmed < capacity:
            # If there is room, confirm them!
            confirmed += 1
        else:
            # If the room is full, put them on the waitlist
            waitlisted += 1
            
    # Determine the final status
    if confirmed == capacity:
        status = "Closed"
    else:
        status = "Open"
        
    print(f"\nConfirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlisted}")
    print(f"Registration Status: {status}")

# --- Taking Inputs ---
event_capacity = int(input("Enter Event Capacity: "))
incoming_registrations = int(input("Enter total number of registrations: "))

manage_registration(event_capacity, incoming_registrations)
```
